```python
from scipy.stats import norm
```

```python
xs = np.linspace(-3, 3)
ys = norm(1, 1).cdf(xs) #mean 1, sd 1
plt.plot(xs, ys)
```

```python
xs = np.linspace(-3, 3)
ys = norm(0, 1).cdf(xs) #mean 0, sd 1
plt.plot(xs, ys)
```

```python
xs = np.linspace(-3, 3)
ys = norm(-1, 1).cdf(xs) #mean -1, sd 1
plt.plot(xs, ys)
```

```python
xs = np.linspace(-3, 3)
ys = norm(1,1).pdf(xs) #mean 0, sd 1
plt.plot(xs, ys)
```
